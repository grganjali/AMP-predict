```python
df_apd3= pd.read_csv("/home/anjali/06/datasets/apd3.csv")
df_camp= pd.read_csv("/home/anjali/06/datasets/camp.csv")
df_lamp= pd.read_csv("/home/anjali/06/datasets/lamp.csv")
df_avppred= pd.read_csv("/home/anjali/06/datasets/avppred.csv")
```

```python
def boxplots(feature= " tiny"):
    
    boxplot_apd3= go.Box(
    name= "dataset_apd3",
    y= df_apd3[feature]
    )
    
    boxplot_camp= go.Box(
    name= "dataset_camp",
    y= df_camp[feature]
    )
    
    boxplot_lamp= go.Box(
    name= "dataset_lamp",
    y= df_lamp[feature]
    )
    
    boxplot_avppred= go.Box(
    name= "dataset_avppred",
    y= df_avppred[feature]
    )
    
    data= [ boxplot_apd3, boxplot_camp, boxplot_lamp, boxplot_avppred]
    iplot(data)
from ipywidgets import interact, fixed
#all we need to specify here is what function we need to interact with
interact(boxplots, feature= df_apd3.columns)    
```
